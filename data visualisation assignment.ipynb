```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from mpl_toolkits.mplot3d import Axes3D
```

```python
iris=sns.load_dataset('iris')
iris.head()
```

```python
fig =plt.figure(figsize=(4,4))
ax=fig.add_subplot(111,projection='3d')     
ax.scatter(iris['sepal_length'],iris['sepal_width'],iris['petal_length'],s=40,)
```

```python
species_labels=np.random.choice([0,1,2],size=150)
fig =plt.figure(figsize=(8,8))
ax=fig.add_subplot(111,projection='3d')    
scat_plot=ax.scatter(iris['sepal_length'],iris['sepal_width'],iris['petal_length'],s=60,c=species_labels)
cb=plt.colorbar(scat_plot,pad=0.2)
cb.set_ticks([0,1,2])
cb.set_ticklabels(["setosa","versicolor","virginica"])
ax.set_title("iris dataset")
ax.set_xlabel("sepal_length")
ax.set_ylabel("sepal_width")
ax.set_zlabel("petal_length")
```
